## Main Function

```python
# Lab Session 1. Regular Expressions
# López Reyes José Roberto. 7CM2.

import re

def color_matched_parts(text, pattern, color_code='\033[91m', reset_code='\033[0m'):
    """
    Resalta las partes de un texto que coinciden con un patrón regex dado, usando un color específico.
    Maneja múltiples coincidencias que no se superponen dentro del mismo texto.
    """
    # Encuentra todas las coincidencias que no se superponen
    matches = []
    for m in re.finditer(pattern, text):
        matches.append((m.start(), m.end()))

    if not matches:
        return text  # Si no hay coincidencia, devuelve el texto original

    # Construye el texto coloreado insertando códigos de color alrededor de las coincidencias
    colored_string_parts = []
    last_end = 0
    for start, end in matches:
        # Agrega la parte del texto que está antes de la coincidencia actual
        colored_string_parts.append(text[last_end:start])
        # Agrega la parte que coincidió, rodeada por los códigos de color
        colored_string_parts.append(color_code)
        colored_string_parts.append(text[start:end])
        colored_string_parts.append(reset_code)
        last_end = end

    # Agrega cualquier parte restante del texto después de la última coincidencia
    colored_string_parts.append(text[last_end:])

    return "".join(colored_string_parts)
```

## An Example

```python
Match = ['wazzzzzup', 'wazzzup']
Skip = ['wazup']
myRegEx = r"..zzz.."

print("Match:")
for s in Match:
    colored_s = color_matched_parts(s, myRegEx)
    print(f"{s} : {colored_s}")

print("\nSkip:")
for s in Skip:
    print(f"{s} : {s}")
```

## Another Example

```python
# Este patrón busca nombres de archivos de Godot (terminados en .gd o .tscn)
myRegEx = r"[a-zA-Z0-9_]+\.(?:tscn|gd)"

Match = [
    "Loading level_01.tscn into memory...",
    "The script player_controller.gd has a logic bug.",
    "Missing files: enemy.tscn, bullet.tscn, and ui.gd! Aborting.",
    "level_boss.tscn"
]

Skip = [
    "Loading player_sprite.png instead.",
    "Check the game_design_doc.pdf for the lore.",
    "This string just has the letters gd and tscn floating around.",
    "error_in_script.gds",
    "tscn.txt"
]

print("--- MATCH EXPECTED ---")
for s in Match:
    colored_s = color_matched_parts(s, myRegEx)
    print(f"{s}\n -> {colored_s}\n")

print("--- SKIP EXPECTED ---")
for s in Skip:
    # Ahora la lista Skip pasa por la función para demostrar que el Regex los ignora
    colored_s = color_matched_parts(s, myRegEx)
    print(f"{s}\n -> {colored_s}\n")
```

## Yet Another Example

```python
# Patrón: Empieza con 'c', tiene exactamente 2 caracteres en medio, termina en 'l'
myRegEx = r"c..l"

Match = [
    "cool",
    "cell",
    "curl",
    "That is so cool man",
    "c44l"
]

Skip = [
    "col",      # Solo tiene 1 caracter en medio
    "camel",    # Tiene 3 caracteres en medio
    "cl",       # No tiene caracteres en medio
    "colorful"  # El patrón "c..l" no encaja aquí (sería c, o, l... espera, "colo" no termina en l!)
]

print("--- MATCH EXPECTED ---")
for s in Match:
    colored_s = color_matched_parts(s, myRegEx)
    print(f"{s} : {colored_s}")

print("\n--- SKIP EXPECTED ---")
for s in Skip:
    colored_s = color_matched_parts(s, myRegEx)
    print(f"{s} : {colored_s}")
```
