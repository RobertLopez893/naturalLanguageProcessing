```python
# Lab Session 2. Exploring a Text
# López Reyes José Roberto. 7CM2.

# English Version

import urllib.request
import re
from collections import Counter
import nltk
from nltk.corpus import stopwords

# Download the stopwords dictionary
nltk.download('stopwords', quiet=True)

def get_top_100_words():
    # Fetch a sample text (Alice's Adventures in Wonderland from Gutenberg)
    print("Fetching the text...")
    url = "https://www.gutenberg.org/cache/epub/11/pg11.txt"

    try:
        response = urllib.request.urlopen(url)
        text = response.read().decode('utf-8')
    except Exception as e:
        print(f"Oops, couldn't fetch the text: {e}")
        return

    # Clean the text
    # Convert to lowercase and grab anything that looks like a word
    print("Processing and counting words...")
    words = re.findall(r'\b[a-z]+\b', text.lower())

    # Filter out the stop words
    stop_words = set(stopwords.words('english'))

    # We'll also filter out some common Gutenberg boilerplate words
    # and single-letter artifacts just to keep the list clean
    custom_filters = {'gutenberg', 'project', 'tm', 'work', 'any'}
    stop_words = stop_words.union(custom_filters)

    meaningful_words = [word for word in words if word not in stop_words and len(word) > 1]

    # Count them up
    word_counts = Counter(meaningful_words)
    top_100 = word_counts.most_common(100)

    # Print the results nicely
    print("\n--- TOP 100 MOST FREQUENT WORDS ---")
    print(f"{'Rank':<6} | {'Word':<15} | {'Count'}")
    print("-" * 35)

    for rank, (word, count) in enumerate(top_100, 1):
        print(f"{rank:<6} | {word:<15} | {count}")

# Run the function
get_top_100_words()
```
